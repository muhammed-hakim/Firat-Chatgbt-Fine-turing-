# 3. Veri Setinin Yüklenmesi ve Model Yapılandırması
Açıklama: Bu aşamada, üniversite yönetmeliğine göre hazırlanan dataset.json dosyası sisteme dahil edilmektedir. Model olarak, Meta'nın güçlü mimarisi üzerine inşa edilen Llama-3-8B seçilmiştir. Donanım kaynaklarını verimli kullanmak adına BitsAndBytesConfig ile 4-bit kuantizasyon (NF4) yapılandırması uygulanmıştır. Bu sayede, yüksek performanslı bir dil modeli, GPU belleği aşılmadan akademik yanıtlar üretebilecek şekilde yüklenmektedir.

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

!pip install -U bitsandbytes

from datasets import load_dataset
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
import torch


dataset = load_dataset('json', data_files='dataset.json') # Removed 'field="None"'


bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

model_id = "unsloth/llama-3-8b-bnb-4bit"
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(model_id, quantization_config=bnb_config)
```

# 4. PEFT Yapılandırması ve İnce Ayar (Fine-Tuning) Süreci
Açıklama: Bu hücrede, modelin özelleştirilmesi için LoRA (Low-Rank Adaptation) mimarisi kurulmaktadır. Bu teknikle, modelin tüm parametrelerini eğitmek yerine sadece belirli katmanlar (q_proj, k_proj vb.) üzerinde güncelleme yapılarak eğitim süresi ve kaynak kullanımı optimize edilir. SFTTrainer aracılığıyla, üniversiteye özgü soru-cevap formatı modele öğretilmektedir. bfloat16 hassasiyeti ve öğrenme oranı (learning rate) gibi parametreler, modelin verilerden en yüksek doğruluğu elde etmesi için akademik standartlarda ayarlanmıştır.

```python
from trl import SFTTrainer
from transformers import TrainingArguments
from peft import LoraConfig, get_peft_model
import os
os.environ["WANDB_DISABLED"] = "true"

# Define the formatting function
def formatting_function(example):
    # Assuming your dataset examples have 'instruction' and 'output' fields.
    # Adjust this template based on your actual dataset structure and desired output format.
    text = f"### Soru:\n{example['instruction']}\n"
    if 'output' in example and example['output']:
        text += f"### Cevap:\n{example['output']}"
    return text

# Configure PEFT (LoRA)
peft_config = LoraConfig(
    r=16, # Rank
    lora_alpha=32, # Alpha
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj",
                    "gate_proj", "up_proj", "down_proj",],
    lora_dropout=0.05, # Dropout
    bias="none", # Bias type
    task_type="CAUSAL_LM",
)

# Apply PEFT to the model
model = get_peft_model(model, peft_config)

trainer = SFTTrainer(
    model=model,
    train_dataset=dataset['train'],
    # tokenizer=tokenizer, # Removed this line as it's causing TypeError
    formatting_func=formatting_function, # Replaced dataset_text_field
    # max_seq_length=512, # Removed this line as it's causing TypeError
    args=TrainingArguments(
        per_device_train_batch_size=2,
        gradient_accumulation_steps=4,
        warmup_steps=5,
        max_steps=60,
        learning_rate=2e-4,
        bf16=True, # Changed from fp16=False to bf16=True for bfloat16 compatibility
        logging_steps=1,
        output_dir="outputs",
    ),
)

trainer.train()
```

# 5. Eğitim Performans Analizi (Loss Grafiği)
Açıklama: Bu grafik, modelin eğitim süreci boyunca gösterdiği performansı temsil etmektedir. Mavi çizgi, modelin hata payının (Loss) zamanla nasıl azaldığını gösterir. Grafikteki düşüş eğilimi, modelin üniversite yönetmeliği verilerini başarıyla öğrendiğini ve sorulara verdiği yanıtların doğruluğunun arttığını bilimsel olarak kanıtlamaktadır.

```python
import matplotlib.pyplot as plt

# trainer.state.log_history
history = trainer.state.log_history

# Loss (Steps)
steps = [x['step'] for x in history if 'loss' in x]
losses = [x['loss'] for x in history if 'loss' in x]


plt.figure(figsize=(10, 5))
plt.plot(steps, losses, label='Eğitim Kaybı (Training Loss)', color='blue', linewidth=2)
plt.title('Model Eğitim Süreci: Loss Grafiği')
plt.xlabel('Adımlar (Steps)')
plt.ylabel('Hata Payı (Loss)')
plt.legend()
plt.grid(True)
plt.show()
```

# 7. İnteraktif Yanıt Üretimi ve Akış Testi
Açıklama: Bu hücrede, modelin gerçek zamanlı yanıt verme yeteneği test edilmektedir. TextStreamer kullanılarak, modelin cevabı bir kerede değil, kelime kelime ekrana yazdırılması (streaming) sağlanır; bu da kullanıcı deneyimini iyileştiren profesyonel bir yaklaşımdır. Belirlenen akademik soruya verilen yanıt, modelin sadece bilgiyi öğrenmekle kalmadıp، aynı zamanda bu bilgiyi kurumsal bir dil ile nasıl yapılandırdığını göstermektedir.

```python

from transformers import TextStreamer


soru = "Tez konusu önerisi için Enstitüye gönderilmesi zorunlu olan formlar hangileridir?"


prompt = f"      ### Soru :\n\n{soru}\n\n### Cevap :\n\n"


inputs = tokenizer([prompt], return_tensors = "pt").to("cuda")

#  (Generator)
print("\n--- Modelin Cevabı ---\n")
_ = model.generate(**inputs, streamer = TextStreamer(tokenizer), max_new_tokens = 128)

soru = "Yüksek lisans tez jürisi kaç kişiden oluşur?"


prompt = f"      ### Soru :\n\n{soru}\n\n### Cevap :\n\n"


inputs = tokenizer([prompt], return_tensors = "pt").to("cuda")

#  (Generator)
print("\n--- Modelin Cevabı ---\n")
_ = model.generate(**inputs, streamer = TextStreamer(tokenizer), max_new_tokens = 128)
```

# 8. Kullanıcı Arayüzü Oluşturma ve Yayına Alma (Gradio)
Açıklama: Projenin son aşamasında, eğitilen modelin son kullanıcılar tarafından kolayca kullanılabilmesi için Gradio kütüphanesi ile grafiksel bir arayüz tasarlanmıştır. Bu hücredeki kod, karmaşık yapay zeka süreçlerini basit bir "Soru-Cevap" kutusuna dönüştürür. iface.launch(share=True) komutu sayesinde, yerel bir sunucu üzerinden model dış dünyaya açılır ve akademik prosedürler hakkında sorulan sorulara anlık, doğru ve profesyonel yanıtlar veren bir "Akademik Asistan" web uygulaması başlatılmış olur.

```python

!pip install -q gradio

import gradio as gr


def firat_gpt_respond(user_question):
    # JSON
    prompt = f"Soru: {user_question}\nCevap:"

    # (GPU)
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")


    outputs = model.generate(**inputs, max_new_tokens=100, temperature=0.7)


    full_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Cevap:
    response = full_text.split("Cevap:")[-1].strip()
    return response


iface = gr.Interface(
    fn=firat_gpt_respond,
    inputs=gr.Textbox(lines=2, placeholder="Lütfen üniversite ile ilgili sorunuzu buraya yazın..."),
    outputs="text",
    title="Fırat-GPT: Akademik Asistan",
    description="Fırat Üniversitesi Fen Bilimleri Enstitüsü ve Akademik prosedürler hakkında eğitilmiş özel yapay zeka.",
    examples=[["Form-30 nedir?"], ["Tez savunma jürisi kaç kişiden oluşur?"], ["Turnitin benzerlik oranı kaç olmalıdır?"]]
)

# (Public Link)
iface.launch(share=True)
```
